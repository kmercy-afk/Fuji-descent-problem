```python
import numpy as np

csv_path = "C:/Users/Mercy/downloads/mtfuji_data.csv" 

np.set_printoptions(suppress=True) 
fuji = np.loadtxt(csv_path, delimiter=",", skiprows=1)
print(fuji)
```

```python
import matplotlib.pyplot as plt

plt.title("Mt. Fuji")
plt.xlabel("position")
plt.ylabel("elevation[m]")
plt.tick_params(labelsize=15)
plt.plot(fuji[:,0], fuji[:,3])

plt.show()
```

```python
x = fuji[:,0]
y = fuji[:,3]
def gradient(i):
    if i == 0:  # 最初の点では勾配を定義できないので0を返す
        return 0
    return (y[i] - y[i - 1]) / (x[i] - x[i - 1])
#example
gradient(100)
```

```python
def move(i, alpha):
    new_index = fuji[i,3] - alpha * gradient(i)  # 次の地点を計算
    new_index = max(0, round(new_index))  # 負の値を防ぎ、整数に変換
    return new_index
#example
move(299,0.2)
```

```python
def down(i):
  point = []
  for j in range(299-i):
    point.append(move(i+j, 0.2))
  return point
#130
down(130)
```

```python
plt.title("Mt. Fuji from No.136")
plt.xlabel("position")
plt.ylabel("elevation[m]")
plt.tick_params(labelsize=15)
plt.plot(down(136))

plt.show()
```
